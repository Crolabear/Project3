```python
# identify similar pairs
SimArrar=np.array(SimScore2)
SimArrar1=np.array(gg)


x=[]
y=[]
# x, and y are the indice of the title matrix

for i in range(2500):
    for j in range (2500):
        if (SimArrar[i,j]>0.7 and SimArrar[i,j]!=1):
            x.append(i)
            y.append(j)
            
#Obtain the indice that are more than 90% similar.            
                                            
#ArrayTitle=np.concatenate([x,y],axis=0)
#Row=len(ArrayTitle)
#ArrayTitle=ArrayTitle.reshape(Row/2,2)



#xlab=map(lambda z: Names[z],x)
#ylab=map(lambda z: Names[z],y)

```

The titles that we classified as similar share some features in common: one title tend to be very specific and the other title is a broader or generalization of the prevous title: politic and political checking; malefrugalfashion and femalefrugalfashion, etc. 

```python
Titlex22=[]
Titley22=[]

for i in range(len(x)):
    d=x[i]-y[i]
    if d > 2:
        Titlex22.append(x[i])
        Titley22.append(y[i])

#~~~
# 2 .._ titl info
xlab2=map(lambda z: Names[z],Titlex22)
ylab2=map(lambda z: Names[z],Titley22)


labels=map(lambda a,b: a+'\n'+ b, xlab2,ylab2)
plt.subplots_adjust(bottom = 0.1)
plt.scatter(
    Titlex22,Titley22, marker = 'o')
for label, x, y in zip(labels, Titlex22, Titley22):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'),fontsize=8)

plt.suptitle(' Selected subreddits whose titles are over 90% similar')
plt.xlabel('subreddit indice')
plt.show()
```

The title approach was able to give us some idea and potential candidates of similar posts. Let's look into post content to see if we can capture what is not being captured with the title approach. I originally tried two methods to measture content similarities: Inverse IDF and the simple word count. The Inverse IDF approach weights down words that repeatedly appears and weight up the words that appear only occasionally. This may not be the best approaach since what makes a post unique may be the repeated occurance of certain words. Therefore, the Inverse IDF method classified all the posts are similar to each other in term of contents.

The simple word count approach may favor subreddits with more words, but it was able to give an outcome that agrees with title approach.
